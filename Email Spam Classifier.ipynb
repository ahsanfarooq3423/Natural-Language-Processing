```python
data = pd.read_csv('emails.csv')
data.head(10)

#Removing the common 'subject' word in emails text
data['text'] = data['text'].map(lambda x: x.lstrip('Subject:'))
data.head()
```

```python
spam_data = data[data['spam'] == 1]
non_spam_data = data[data['spam'] == 0]
```

```python
raw_spam = spam_data["text"].values
raw_text = ''.join(raw_spam)
len(raw_text)
```

```python
#Cleaning the spam text
cleanText = re.sub('\W+',' ', raw_text )
cleanText_tokenize = nltk.tokenize.word_tokenize(cleanText)
fdist = nltk.FreqDist(cleanText)
cleanText_tokenize= [i for i in cleanText_tokenize if len(i) > 1]
final =[word for word in cleanText_tokenize if word not in stopwords.words('english')]

fdist = nltk.FreqDist(final)
temp = fdist.most_common(3000)
print(temp[0][0])


spam_vocabulary = []
for i in range(len(temp)):
    spam_vocabulary.append(temp[i][0])
    
    
```
